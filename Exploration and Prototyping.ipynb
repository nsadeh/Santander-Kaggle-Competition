```python
# remove dead columns
def is_dead(df, col_name):
    if np.issubdtype(df.loc[:, col_name], np.number):
        return np.std(df.loc[:, col_name].values) == 0
    else:
        return False

dead_cols = list(filter(lambda col: is_dead(training_data_file, col), list(training_data_file.columns)))
print('There are ' + str(len(dead_cols)) + ' empty columns in the dataset')
training_data = training_data_file.drop(dead_cols, axis = 1)
assert len(training_data.columns)+ len(dead_cols) == len(training_data_file.columns) # check
```

```python
# distribution of target
target = training_data.target.values
plt.figure(figsize = (18,12))
x = plt.hist(np.log(target) + 1, bins = 'auto', normed = True)
t = plt.title('Distribution of Logarithm of Response Var')
x = plt.xlabel('Log Response')
y = plt.ylabel('Density')
```
